```python
df2=df.groupby(['DESCRIPTION'],as_index=False).sum()  # it split the description and unit into groups and we will groupby it in new variable df2
```

```python
df2.fillna(0) #will replace null values  with 0
```

```python
df2.UNIT=df2.UNIT.astype(int)  #as we want to plot bar chart so we will convert our float values into int type 
df2
```

```python
import matplotlib.pyplot as plt
```

```python
#bar chart of description vs unit 
decription=df2['DESCRIPTION']
count=df2['UNIT']
fig = plt.figure(figsize =(10, 7))
plt.bar(decription,count)
plt.show()
#intemwise sale 
```
